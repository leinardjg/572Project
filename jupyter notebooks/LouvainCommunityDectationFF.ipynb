```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas

import networkx.algorithms.community as nx_comm
%matplotlib inline
```

```python
#read the edge file with 

G = nx.read_edgelist('./edges.csv',
                     create_using=nx.DiGraph(), delimiter = ', ')
```

```python
# create a collection of louvain community dectectation algorithms iterations

collect_results = []
store_modularity = []

for i in range(100):
    Community_Parts = nx_comm.louvain_communities(G , seed = i)
    m = nx_comm.modularity(G, Community_Parts)
    collect_results.append(Community_Parts)
    store_modularity.append(m)
```

```python
max_Modularity = max(store_modularity)

index = 0
count = 0
for i in range(len(store_modularity)):
    if(store_modularity[i] == max_Modularity):
        index = i
        count += 1
```

```python
final_ComDetect = collect_results[index]
avg = np.mean(store_modularity)
```

```python
print("Maximum Modularity FF:" , max_Modularity)
print("Index of the max mod FF: ", index)
print("How many max mod exist FF:",count)
print("Number of partitions",len(final_ComDetect))
```

```python
NODES_FILE = './nodes_repo_community.csv'
EDGES_FILE = './edges.csv'
OUTPUT_NODES_FILE = './nodes_repo_Com_FF&Forks.csv'

G = nx.read_edgelist(EDGES_FILE, create_using=nx.DiGraph(), delimiter = ', ')
```

```python
communities =  final_ComDetect
# record of all nodes, key is user login
record = {}

nodesFile = open(NODES_FILE, 'r')

# skip headers
#nodesFile.readline()

# store nodes in record
for line in nodesFile:
    toRecord = line.split(", ")
    user = toRecord[0].strip()
    repo = toRecord[1].strip()
    follower_community = toRecord[2].strip()
    fork_community = -1

    record[toRecord[0]] = [repo, follower_community, fork_community]

# record community
communityNumber = 0

for community in communities:
    for contributor in community:
        record[contributor][2] = communityNumber
    communityNumber += 1

newNodesFile = open(OUTPUT_NODES_FILE, 'w+')

for contributor in record:
    newNodesFile.write(f"{contributor}, {record[contributor][0]}, {record[contributor][1]}, {record[contributor][2]}\n")

nodesFile.close()
newNodesFile.close()
```
